# ANN Model Training for Movie Recommendations

This notebook demonstrates training the Artificial Neural Network (ANN) collaborative filtering model.

```python
import sys
import os
sys.path.append('../backend')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from models.ann_model import ANNCollaborativeFilteringModel
from utils import load_pickle

print("Libraries imported successfully!")
print(f"TensorFlow version: {tf.__version__}")
```

## Load Preprocessed Data

```python
# Load preprocessed data
data_path = '../backend/data/preprocessed.pkl'
data = load_pickle(data_path)

ratings_df = data['ratings_normalized']
user_to_idx = data['user_to_idx']
movie_to_idx = data['movie_to_idx']

print(f"Loaded {len(ratings_df)} ratings")
print(f"Users: {len(user_to_idx)}, Movies: {len(movie_to_idx)}")
```

## Initialize and Train ANN Model

```python
# Initialize ANN model
model = ANNCollaborativeFilteringModel(
    n_users=len(user_to_idx),
    n_movies=len(movie_to_idx),
    embedding_dim=32,
    hidden_dims=[64, 32]
)

# Prepare training data
X_train, X_val, y_train, y_val = model.prepare_training_data(
    ratings_df, user_to_idx, movie_to_idx
)

# Train model
history = model.train(
    X_train, X_val, y_train, y_val,
    epochs=20,
    batch_size=256
)

print("Training completed!")
```

## Evaluate Model

```python
# Evaluate model
metrics = model.evaluate_model(X_val, y_val)
print(f"RMSE: {metrics['rmse']:.4f}")
print(f"MAE: {metrics['mae']:.4f}")

# Plot training history
model.plot_training_history()
```

## Save Model

```python
# Save trained model
model_path = '../backend/models/ann_model'
model.save_model(model_path)
print(f"Model saved to {model_path}")
```
